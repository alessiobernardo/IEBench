```python
import pandas as pd
import numpy as np
```

```python
names = ["Elec", "PAKDD", "KDDCup"]

rates = {}

dfs = {}


for name in names:
    print(name)
    dfs[name] = pd.read_csv(f"{name}.csv", header = None)

for name,df in dfs.items():
    
    w_len= 5000
    window = np.zeros(shape=(w_len,))
    i = 0
    IR = []
    for index, row in df.iterrows():
        window[i] = row[df.columns[-1]]

        ones = len(window[window==1])
        zeros = len(window[window==0])

        if index < w_len:
            IR.append(ones/(index + 1))
        else: 
            IR.append(ones/w_len)
        if index % w_len == 0:
            print(index)
        i = (i + 1) % w_len
    rates[name] = IR

    
    
```

```python
import csv
with open('Real_Imbalance_rates_w5000.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    writer.writerow(["dataset","instances","ones%"])
    for ds, array in rates.items():
        for i,r in enumerate(array):
            writer.writerow([ds,i+1,r])
```
