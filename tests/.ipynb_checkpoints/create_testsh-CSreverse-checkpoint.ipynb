```python
n_exp = 10

algorithms = {
    "old" : [ "CSMOTE"]#, "RebalanceStream","CSMOTE",
    #"new" : ["improved_OOB","improved_UOB","OOB","UOB","WEOB1","WEOB2","RebalanceStream"] 
}

datasets = {
    #"real" : ["Elec","PAKDD", "KDDCup"],
    "art1" : ["sea_", "sine_","seaG_","sineG_"],
    "art2" : ["incremental","sudden", "periodic"]
}

directories = {
    "art1" : ["P(y)","P(x_y)","P(y_x)"],
    "res_art1" : ["P\\(y\\)","P\\(x_y\\)","P\\(y_x\\)"],
    #"name_art1" : ["Py","Px_y","Py_x"],
    
    #

    "art2" : ["appearing-minority","disappearing-minority","jitter","clusters-movement","appearing-clusters",
              "splitting-clusters", "borderline", "shapeshift", "minority-share"]
}



seeds = [0,11,13,17,23,30,39,912,1237,7591]
cnt = 0
options = [1,2,3,4]
test = open("TestWsmallCSrev.txt","a")

for i in range(10,6,-1):
    test.write(f"date\n")
    for o in options:
        
        

        for alg in algorithms["old"]:
            for di,d in enumerate(directories["art1"]): 
                
                for data in datasets["art1"]:
                    test.write(f"echo '{i} {o} {alg} {d} {data[:2]}'\n")

                    l = f'(meta.{alg})'
                    if "ESOS" in alg:
                        l = f'(meta.{alg} -c (OS_ELM -b 100 -i 100 -p) -e (WELM -p -i 100))'

                    elif "CSMOTE" in alg:
                        l = f'(meta.{alg} -l trees.HoeffdingAdaptiveTree -w 100000 -r {seeds[i-1]})'
                    elif "mpr" in alg:
                        alg1 = ''.join(alg)
                        alg2 = alg.replace('_', '').replace("i","I")
                        alg = alg1
                        l = f'(meta.{alg2} -r {seeds[i-1]})'
                    elif "Reb" in alg:
                        l = f'(meta.{alg} -l trees.HoeffdingAdaptiveTree -r {seeds[i-1]})'

                    cnt += 1
                    test.write(f"java -Xmx14g -Xss50M -cp moa.jar -javaagent:sizeofag-1.0.4.jar moa.DoTask '"
                          f'EvaluatePrequential -l {l} -s '
                          f'(ArffFileStream -f ../datasets/SeaSine/{d}/{data}{str(o)},{str(10-o)}.arff) -e '
                          f"(WindowFixedClassificationPerformanceEvaluator -w 5000 -r -f -g) -i -1 -f 5000' "
                          f"1> resultsWsmall/{directories['res_art1'][di]}/{data}{str(o)}_{str(10-o)}/{alg}/{str(i)}_err.csv 2> resultsWsmall/{directories['res_art1'][di]}/{data}{str(o)}_{str(10-o)}/{alg}/{str(i)}.csv\n") 
        for alg in algorithms["old"]:
            for d in directories["art2"] : #appearing-minority
                for data in datasets["art2"]:
                    test.write(f"echo {i} {o} {alg} {d} {data}\n")
                    #for i in range(1,n_exp+1):


                    l = f'(meta.{alg})'
                    if "ESOS" in alg:
                        l = f'(meta.{alg} -c (OS_ELM -b 100 -i 100 -p) -e (WELM -p -i 100))'

                    elif "CSMOTE" in alg:
                        l = f'(meta.{alg} -l trees.HoeffdingAdaptiveTree -w 100000 -r {seeds[i-1]})'

                    elif "mpr" in alg:
                        alg1 = ''.join(alg)
                        alg2 = alg.replace('_', '').replace("i","I")
                        alg = alg1
                        l = f'(meta.{alg2})'
                    elif "Reb" in alg:
                        l = f'(meta.{alg} -l trees.HoeffdingAdaptiveTree)'


                    cnt += 1
                    test.write(f"java -Xmx14g -Xss50M -cp moa.jar -javaagent:sizeofag-1.0.4.jar moa.DoTask '"
                          f'EvaluatePrequential -l {l} -s '
                          f'(ArffFileStream -f ./drifts_arff/{d}-{data}-{o}-{i-1}.arff) -e '
                          f"(WindowFixedClassificationPerformanceEvaluator -w 5000 -r -f -g) -i 100000 -t 500 -f 5000' 1> resultsWsmall/Drifts/{d}/{str(o)}/{alg}/{data}_{str(i)}_err.csv 2> resultsWsmall/Drifts/{d}/{str(o)}/{alg}/{data}_{str(i)}.csv\n")


test.close()
print(cnt)
```
